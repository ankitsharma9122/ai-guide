# Transformer 

## What is Transformer?
Transformer is a deep learning model used in AI systems like GPT and BERT.
It processes text by understanding relationships between words using "attention".

---

# RNN & LSTM (Sequential Models)

## What is RNN?

RNN (Recurrent Neural Network) is a neural network designed for sequence data.

It processes input one step at a time and keeps memory of previous steps.

---

## How RNN works

Example: "I love AI"

Step-by-step:
1. Read "I"
2. Pass hidden state → next step
3. Read "love"
4. Pass hidden state → next step
5. Read "AI"

So each step depends on previous one.

---

## Problem with RNN

1. Slow (no parallel processing)
2. Vanishing Gradient Problem
   - Long sentences → forget early words
3. Cannot handle long context

---

## What is LSTM?

LSTM (Long Short-Term Memory) is an improved version of RNN.

It is designed to remember long-term dependencies.

---

## How LSTM solves RNN problems

LSTM uses gates:

1. Forget Gate → what to forget
2. Input Gate → what to store
3. Output Gate → what to output

---

## LSTM Advantages

- Remembers long sequences better
- Reduces vanishing gradient problem

---

## Still Problems in LSTM

- Still sequential (slow)
- Cannot fully parallelize
- Complex architecture
- Training takes time

---

## Simple Summary

RNN/LSTM process text step-by-step → slow and limited for long context.

## Why Transformer was needed (Problem it solved)


### ❌ Before Transformer (RNN / LSTM problems)
- Text processed one-by-one (slow)
- Hard to remember long sentences
- Poor performance on long context
- Cannot parallelize (training slow)

### ✅ Transformer solved:
- Processes all words at once (fast)
- Understands long-range relationships
- Uses parallel computation
- Better accuracy on language tasks

---

## Key Idea: Attention

Instead of reading sequentially, Transformer checks:
"Which words are important for this word?"

Example:
Sentence: "The bank is near the river"
- "bank" focuses on "river" → meaning = river bank

---

## Main Components

1. Embedding
- Converts words → numbers (vectors)

2. Positional Encoding
- Adds word order info

3. Self-Attention ⭐
- Each word looks at all other words

4. Multi-Head Attention
- Multiple attention layers to learn different patterns

5. Feed Forward Network
- Simple neural network for processing

6. Layer Norm + Residual
- Helps stable training

---

## Encoder vs Decoder

- Encoder → Understand input (BERT)
- Decoder → Generate output (GPT)

---

## Flow

Input Sentence
↓
Tokenization
↓
All tokens processed at once
↓
Self-Attention (matrix computation)
↓
Output


## IN RNN/LSTM
processed one by one like chaining

h1 → h2 → h3 → h4 (chain)

## In Attention based model(Transformer)
Each word looks at all words simultaneously.

---

## Simple Summary

Transformer = Attention-based model that understands full sentence at once instead of step-by-step.

# Tokens & Tokenization

## What is a Token?

Token = smallest unit of text used by AI model.

Important:
Token is NOT always a word.

---

## Examples

"Hello world"
→ ["Hello", "world"]

"playing"
→ ["play", "ing"]

"ChatGPT"
→ ["Chat", "G", "PT"]

---

## Why Tokens are important

- Model reads tokens, not text
- Cost depends on tokens
- Input limit depends on tokens

---

## What is Tokenization?

Tokenization = converting text into tokens.

---

## Why Tokenization was needed (Problem it solved)

### ❌ Without Tokenization
- Computer cannot understand raw text
- Words are not numerical
- Huge vocabulary problem

### ❌ Word-level issues
- Too many words (memory issue)
- Unknown words problem

### ❌ Character-level issues
- Too long sequences
- Slow processing

### ✅ Tokenization solved:
- Converts text → numbers
- Handles unknown words
- Reduces vocabulary size
- Efficient processing

## Example :
["I", "am", "playing", "cricket"]

Now unkown like
"I am playzzing cricket"

solution :

"I am playing cricket"
↓
["I", "am", "play", "ing", "cricket"]

and 

["I", "am", "play", "zz", "ing", "cricket"]
---

## Types of Tokenization

1. Word Tokenization
"I love AI" → ["I", "love", "AI"]

2. Character Tokenization
"AI" → ["A", "I"]

3. Subword Tokenization (BEST)
"playing" → ["play", "ing"]

# Multi-shot Prompting (Few-shot Learning)

## What is Multi-shot Prompting?

Providing multiple examples in prompt to guide model output.

---

## Why it is needed (Problem it solved)

### ❌ Without examples (Zero-shot)
- Model may misunderstand task
- Output inconsistent
- Format issues

### ❌ One-shot
- Not enough learning signal

### ✅ Multi-shot solves:
- Gives clear pattern
- Improves accuracy
- Better formatting
- Works like teaching by examples

---

## Types of Prompting

1. Zero-shot
No example

2. One-shot
One example

3. Multi-shot (Few-shot) ⭐
Multiple examples

---

## Example

### Zero-shot
Translate English to French:
"Hello"

---

### Multi-shot

Translate English to French:

Hello → Bonjour  
Thank you → Merci  
Good morning → Bonjour  

Now:
Good night →

---

## How it works

Model learns pattern from examples and applies to new input.

---

## Structure

Instruction
+
Examples (input → output)
+
New input

---

## Use Cases

- Translation
- Classification
- Code generation
- Data formatting

---

## Simple Summary

Multi-shot prompting = teaching AI using examples inside the prompt.

# Context Window (LLM Concept)

## What is Context Window?

Context window is the maximum number of tokens a model can read or remember at one time.

 In simple words:
It is the "memory limit" of the model.

---

## Example

If a model has 8K context window:

- It can read up to 8000 tokens at once
- Includes:
  - User input
  - Previous conversation
  - System prompt

---

## Real Example

Conversation:

User: "My name is Ankit"
User: "I live in Bihar"
User: "What is my name?"

If within context window:
→ Model answers: "Ankit" 

If outside context window:
→ Model forgets 

---

## Why Context Window is Important (Problem it solves)

### ❌ Without Context Window Limit
- Infinite memory needed (not possible)
- Computation becomes very expensive
- Slower processing

### ✅ Context Window solves:
- Limits memory usage
- Keeps processing efficient
- Enables handling long conversations (within limit)
