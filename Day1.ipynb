#  System Prompt vs User Prompt 

---

##  Definition 

* **System Prompt** → sets behavior of AI
* **User Prompt** → actual question asked to AI

---

##  Easy Understanding

1. System Prompt = **Instructions / Role**
2. User Prompt = **Task / Question**

---

##  Real-life Example

System Prompt:
"You are a doctor"

User Prompt:
"I have a headache"

AI will answer like a doctor

---

##  Code Example

```python
system_prompt = "You are a senior backend engineer"
user_prompt = "Explain caching"

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=f"{system_prompt}\n{user_prompt}"
)

print(response.text)
```

---

##  Why System Prompt is Important

Same question → different answers

Example:

User Prompt: "Explain API"

| System Prompt          | Output                |
| ---------------------- | --------------------- |
| "You are a teacher"    | Simple explanation    |
| "You are an architect" | Deep technical answer |

---

##  Interview Answer

 "System prompt defines how the AI should behave, like its role, tone, and rules.
User prompt is the actual input or question given by the user.
Together, they control the AI’s output."

---

## Best Practice

✔ Keep system prompt fixed
✔ Change user prompt dynamically

---

## Common Mistake

1. Mixing both in one string without structure
2. Not controlling tone → inconsistent output

---

## Pro Tip

"System prompts are used to control LLM behavior and are critical in production systems like chatbots, copilots, and AI assistants."

---

## Final Shortcut

1. System Prompt = **How AI behaves**
2. User Prompt = **What AI answers**
---


````python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

system_prompt = "You are a good backend engineer. Give clear and optimized explanations in 1 line"
user_prompt = "What is caching?"

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=[
        {
            "role": "system", # behavior set karta hai
            "parts": [{"text": system_prompt}]
        },
        {
            "role": "user", # actual question
            "parts": [{"text": user_prompt}]
        }
    ]
)

print(response.text)
````
