# Distillation

it is a technique where a smaller model is trained to mimic a larger model's behavior to achieve efficiency with minimal performance loss.

Distillation = making a smaller model learn from a bigger model

Think like:

Big model (teacher) = GPT-4

Small model (student) = lightweight AI

Student learns how teacher thinks → becomes faster & cheaper


*Teacher model output
teacher_output = "Paris is the capital of France."

* Train student model using teacher output
student_model.train(input="What is capital of France?", output=teacher_output)

# Chain of Thought

Chain of Thought prompting improves model reasoning by encouraging step-by-step explanation before producing the final answer.


prompt = """
Solve step by step:
If a train travels 60 km in 1 hour,
how much in 3 hours?
"""

| Feature | Distillation      | Chain of Thought      |
| ------- | ----------------- | --------------------- |
| Purpose | Smaller model     | Better reasoning      |
| Used in | Training          | Prompting             |
| Goal    | Efficiency        | Accuracy              |
| Example | Teacher → Student | Step-by-step thinking |
